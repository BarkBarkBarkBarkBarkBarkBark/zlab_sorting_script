```python
import numpy as np
import matplotlib.pyplot as plt
from probeinterface import Probe

n = 8
pitch_um = 500.0

# 2x4 grid in XY
pos2 = np.c_[ (np.arange(n) // 4) * pitch_um,
              (np.arange(n) %  4) * pitch_um ]

probe = Probe(ndim=2, si_units="um")
probe.set_contacts(pos2, shapes="circle", shape_params={"radius": 5.0})

probe3 = probe.to_3d(axes="xy")  # z=0 plane

pos3 = probe3.contact_positions
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")
ax.scatter(pos3[:, 0], pos3[:, 1], pos3[:, 2])
plt.show()
```
